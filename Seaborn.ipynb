# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Load dataset
iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

# Create scatter plot
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris)
# Add labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot of Sepal Length vs Petal Length")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]
category = ['A', 'A', 'B', 'B', 'A']

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'category': category})
print(data)

# Plot with hue
sns.scatterplot(data=data, x='x', y='y', hue='category')

# Show the plot
plt.show()
```

# Heat map

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
df1=iris.drop('species', axis='columns')
# Compute correlation matrix
corr = df1.corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap="viridis")

# Add title
plt.title("Correlation Matrix of Iris Dataset")

# Show plot
plt.show()
```

# Violin Plot

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('fifa_data.csv')
df
```

```python
df.head()
```

#### 1.Which country has the most number of players 

```python
df['Nationality'].value_counts().idxmax()
```

#### 2.Plot a bar chart of 5 top countries with most number of players

```python
top_countries=df['Nationality'].value_counts().nlargest(5)
top_countries
```

```python
sns.barplot(x=top_countries.index,y=top_countries.values)
plt.xlabel('Countries')
plt.ylabel('No. of Players')
plt.title('5 top countries with most number of players')
plt.show()
```

```python
df['Release Clause'] = df['Release Clause'].str.replace('', '')  # Remove Euro symbol

df['Release Clause'] = df['Release Clause'].str.replace('M', 'e+6')  # Convert M to scientific notation

df['Release Clause'] = df['Release Clause'].str.replace('K', 'e+3')  # Convert K to scientific notation

df['Release Clause'] = pd.to_numeric(df['Release Clause'])
df['Release Clause']
```

#### 3.Which player has the highest salary?

```python
df['Wage']
```

```python
df['Wage'] = df['Wage'].str.replace('', '')
df['Wage'] = df['Wage'].str.replace('K', 'e+3')
df['Wage'] = pd.to_numeric(df['Wage'])
df['Wage']
```

```python
highest_salary_employee=df['Wage'].idxmax()
print(df.loc[(highest_salary_employee,['Name','Wage'])])
```

#### 4.Plot a histogram to get the salary range of the players.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Wage'], bins=20)
plt.xlabel('Wage')
plt.ylabel('Frequency')
plt.title('Salary Range of Players')
plt.show()
```

#### 5.Who is the tallest player in the fifa players?

#### 6.Which club has most number of players?

```python
club=df['Club'].value_counts().idxmax()
club
```

#### 7.Which foot is most preferred by the players?Draw a bar chart for preferred foot

```python
foot_counts = df['Preferred Foot'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=foot_counts.index, y=foot_counts.values)
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.title('Preferred Foot of Players')
plt.show()
```
